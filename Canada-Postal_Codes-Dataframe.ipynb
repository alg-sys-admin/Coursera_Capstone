#### UsePandas library built-in to scrape tabular data from html page

```python
import pandas as pd 

DF = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", header=0)

print(DF[0])
```

#### Grab only needed Columns and Rows

```python
Codes_DF = DF[0].copy()
```

#### User iterate to validate data from table on web page matches

```python
for (index_label, row_series) in Codes_DF.iterrows():
   print('Row: ', index_label, 'Data: ', row_series.values)
   #print('Data : ', row_series.values)
```

```python
for namedTuple in Codes_DF.itertuples():
   #Print row contents inside the named tuple
   print(namedTuple)
```

#### Copy data with assigned Boroughs into new DataFrame

```python
Assigned_DF = Codes_DF[(Codes_DF['Borough'] != 'Not assigned') ].copy()
```

#### Quick check using head

```python
Assigned_DF.head(10)
```

#### Find "Not assigned" Neighbourhoods

```python
Assigned_DF[Assigned_DF["Neighbourhood"] == "Not assigned"] 
```

#### Replace "Not assigned" in Neighbourhood with Borough

```python
Assigned_DF["Neighbourhood"].replace("Not assigned", Assigned_DF.Borough, inplace=True)
```

#### Search again for "Not assigned" Neighbourhoods

```python
Assigned_DF[Assigned_DF["Neighbourhood"] == "Not assigned"]
```

#### Verify Change

```python
Assigned_DF.iloc[6]
```

#### Since this is a small dataset, iterate through dataframe to confirm nothing was missed.

```python
for (index_label, row_series) in Assigned_DF.iterrows():
   print('Row: ', index_label, 'Data: ', row_series.values)
```

#### Group By Postcode and join Neighbourhoods on same row 

```python
Postcodes_DF = Assigned_DF.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(','.join).reset_index()
```
